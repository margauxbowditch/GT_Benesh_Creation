```python
# Preprocess for ExtDerriere Skeletons - Initially done with deleting skeletons:
import os
import json
import pandas as pd

skeleton_dir = "./SkeletonXDataUnnormalized/0-ExtDerriereOnRight"

def deleteFilesNoSkeletons(dir):
    no_skel_counter = 0
    for sfile in os.listdir(dir):
        spath = skeleton_dir + "/" + sfile
        skel_json = pd.read_json(spath)
        if(len(skel_json.people) == 0):
            #TO DO: Delete this file:
            os.remove(spath)
            print(f"No skeleton for {sfile} - it is removed now ")
            no_skel_counter +=1
    print(f"Number of files with no skeletons deleted: {no_skel_counter}")
deleteFilesNoSkeletons(skeleton_dir)
```
